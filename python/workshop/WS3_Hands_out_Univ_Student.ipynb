#### ‚ùì If-Else Statements

````python
# ‚ùì If-Else Statements
# Used for conditional execution.
x = 5

if x > 5:
  print("x is greater than 5")
elif x == 5:
  print("x is equal to 5")
else:
  print("x is less than 5")
````

````python
# Check if a key exists in the dictionary
if 'job' in my_dict:
  print("Job exists in the dictionary.")
else:
  print("Job does not exist in the dictionary.")
````

````python
# Check a condition based on a list
my_list = [1, 2, 3, 4, 5] # len(my_list) = 5
if len(my_list) > 3:
  print("The list has more than 3 elements.")
else:
  print("The list has 3 or fewer elements.")
````

````python
# Combining conditions with logical operators
x = 6
y = 26
if x > 5 and y < 25:
  print("x is greater than 5 and y is less than 25.")
elif x > 5 or y > 25:
  print("x is greater than 5 or y is greater than 25.")
else:
  print("Neither condition is met.")
````

````python
# Nested if-else
z = 25
if z > 10:
  if z < 20:
    print("z is between 10 and 20.")
  else:
    print("z is greater than or equal to 20.")
else:
  print("z is less than or equal to 10.")
````

````python
def analyze_dictionary(data_dict):
  """
  Analyzes a dictionary, demonstrating loops and conditional logic.

  Args:
    data_dict: A dictionary to analyze.
  """
  print("Analyzing dictionary:")
  for key, value in data_dict.items():
    print(f"Key: {key}, Value: {value}")

    # Use if-else to check value types or conditions
    if isinstance(value, int):
      if value > 50:
        print("  Value is an integer greater than 50")
      else:
        print("  Value is an integer less than or equal to 50")

    elif isinstance(value, str):
      if len(value) > 10:
        print("  Value is a string longer than 10 characters")
    else:
      print("  Value is of another type")
````

````python
# Example usage of the analyze_dictionary function
my_extended_dict = {'name': 'Charlie', 'score': 75, 'city': 'Los Angeles', 'is_active': True}
analyze_dictionary(my_extended_dict)
````

````python
# Function to process a list, demonstrating loops and conditional logic

def process_list(input_list):
  """
  Processes a list, demonstrating loops and conditional logic.

  Args:
    input_list: A list of numbers to process.
  """
  print("Processing list:")
  total_sum = 0
  for item in input_list:
    print(f"Processing item: {item}")
    total_sum += item
    print(f' Total sum: {total_sum}')

    # Use if-else to check conditions on list items
    if item % 2 == 0:
      print("  Item is even")
    else:
      print("  Item is odd")
    print("#-------------------------------------")

  print(f"Total sum of list items: {total_sum}")
````

````python
# Example list
my_numbers = [10, 5, 22, 15, 8]
````

````python
# Use a for loop to iterate and call the function for each item if needed
print("Iterating through numbers and processing:")
for number in my_numbers:
  # Here we just print, but you could call another function
  print(f"Outer loop processing: {number}")

# Call the process_list function
print("")
process_list(my_numbers)
````

### üß† Step 11: Interactive Filtering by Species

````python
def filter_species(species):
    display(iris[iris['species'] == species].describe())

widgets.interact(filter_species, species=iris['species'].cat.categories);
````

### üìà Step 12: Interactive Scatter Plot

````python
def interactive_scatter(x_axis='sepal_length', y_axis='sepal_width', color='species'):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=x_axis, y=y_axis, hue=color, data=iris, s=50)
    plt.title(f'Scatter Plot of {y_axis} vs {x_axis}')
    plt.xlabel(x_axis.replace('_', ' ').title())
    plt.ylabel(y_axis.replace('_', ' ').title())
    plt.show()

features = iris.select_dtypes(include = np.number).columns.tolist()
colors = features + ['species']

widgets.interact(interactive_scatter,
                 x_axis=features,
                 y_axis=features,
                 color=colors);
````
